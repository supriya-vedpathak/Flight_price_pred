#### 9.duration

```python
sns.boxplot(df["duration"])
```

```python
q1=np.quantile(df["duration"],0.25)
q3=np.quantile(df["duration"],0.75)
eiqr = 1.5* (q3 - q1)
lower_tail = q1 - eiqr
upper_tail = q3 + eiqr
print(lower_tail , upper_tail)
```

```python
df["duration"]=np.where(df["duration"]>upper_tail,upper_tail,df["duration"])
sns.boxplot(df["duration"])
```

### Model Training

```python
x = df.drop(["price","flight","Unnamed: 0"], axis=1)
y = df["price"]
```

```python
xtr,xts,ytr,yts=train_test_split(x,y,test_size=0.2,random_state=12)
```

```python
linreg=LinearRegression()
linreg.fit(xtr,ytr)
```

```python
yptr = linreg.predict(xtr)
r2_score(ytr,yptr)
```

```python
ypts = linreg.predict(xts)
r2_score(yts,ypts)
```

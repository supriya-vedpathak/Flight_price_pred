### Model Training

```python
x = df.drop(["price","flight","Unnamed: 0"], axis=1)
y = df["price"]
```

```python
x.rename({'class':'Class'}, axis = 1, inplace = True)
```

```python
x.sample(1)
```

```python
xtr,xts,ytr,yts=train_test_split(x,y,test_size=0.2,random_state=12)
```

```python
linreg=LinearRegression()
linreg.fit(xtr,ytr)
```

```python
yptr = linreg.predict(xtr)
r2_score(ytr,yptr)
```

```python
ypts = linreg.predict(xts)
r2_score(yts,ypts)
```
